**1. Faça a leitura de uma imagem 2D colorida qualquer e apresente as
informações listadas abaixo por meio da função print():**

- altura,
- largura,
- canais de cor,
- tipo de dado,
- tom de cinza máximo,
- tom de cinza médio,
- tom de cinza mínimo.

```python
import cv2
from matplotlib import pyplot as plt

# Carregando a imagem
image_path = '/img/C01_61136_Birds_finalist_La-Dama-blanca_Stanislao-Basileo_ITA.jpg'
image = cv2.imread(image_path)

# Convertendo BGR para RGB (para correta exibição com o matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Exibindo a imagem usando o matplotlib
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Verifica se a imagem foi carregada corretamente
if image_rgb is not None:
    # Obtém as dimensões da imagem
    altura, largura, canais = image_rgb.shape

    # Obtém as informações sobre os canais de cor
    canais_de_cor = cv2.split(image_rgb)

    # Obtém o tipo de dado da imagem
    tipo_de_dado = image_rgb.dtype

    # Converte a imagem para escala de cinza para calcular as estatísticas de intensidade
    imagem_escala_cinza = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)

    # Calcula estatísticas de intensidade
    tom_cinza_maximo = imagem_escala_cinza.max()
    tom_cinza_medio = imagem_escala_cinza.mean()
    tom_cinza_minimo = imagem_escala_cinza.min()

    # Imprime as informações
    print(f"Altura: {altura} px")
    print(f"Largura: {largura} px")
    print("Canais de Cor:", canais)
    print("Tipo de Dado:", tipo_de_dado)
    print("Tom de Cinza Máximo:", tom_cinza_maximo)
    print("Tom de Cinza Médio:", tom_cinza_medio)
    print("Tom de Cinza Mínimo:", tom_cinza_minimo)
else:
    print("Falha ao carregar a imagem.")
```

**2. Faça a leitura de uma imagem de entrada 2D colorida qualquer e aplique o filtro da média, mediana e gaussiano na imagem de entrada.**

**Mostre os resultados de cada filtro por meio da biblioteca `matplotlib`.**

```python
import cv2
from matplotlib import pyplot as plt

# Carregando a imagem
image_path = '/img/C02_66687_Mammals_-finalist.jpg'
image = cv2.imread(image_path)

# Convertendo BGR para RGB (para correta exibição com o matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Exibindo a imagem usando o matplotlib
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Configurando o tamanho das figuras
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Exibindo a imagem original
axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title('Original')
axes[0, 0].axis('off')

# Aplicando o filtro da média
image_media = cv2.blur(image, (5, 5))
axes[0, 1].imshow(image_media)
axes[0, 1].set_title('Filtro da Média')
axes[0, 1].axis('off')

# Aplicando o filtro da mediana
image_mediana = cv2.medianBlur(image, 5)
axes[1, 0].imshow(image_mediana)
axes[1, 0].set_title('Filtro da Mediana')
axes[1, 0].axis('off')

# Aplicando o filtro gaussiano
image_gaussiano = cv2.GaussianBlur(image, (5, 5), 0)
axes[1, 1].imshow(image_gaussiano)
axes[1, 1].set_title('Filtro Gaussiano')
axes[1, 1].axis('off')

# Ajustando o layout para evitar sobreposição
plt.tight_layout()

plt.show()
```
